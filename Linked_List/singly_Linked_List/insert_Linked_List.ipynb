```python
class Node:
    """A class to represent a node in the linked list."""
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    """A class to represent a linked list."""
    def __init__(self):
        self.head = None


    def insert_at_beginning(self, data):
        """Inserts a new node at the beginning of the linked list."""
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    
    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node


    def insert_at_middle(self, data, index):
        new_node = Node(data)
        if index == 0:
            new_node.next = self.head
            self.head = new_node

        else:
            temp_node = self.head 
            for _ in range(index - 1):
                temp_node = temp_node.next    
            new_node.next = temp_node.next
            temp_node.next = new_node

    # search in a Linked List
    def search(self, target):
        current = self.head
        while current is not None:
            if current.data == target:
                return True
            current = current.next
        return False



    def update_value(self, index, value):
        if index < -1:
            return None
        
        current = self.head
        for _ in range(index):
            current = current.next
        if current:
            current.data = value
            return True
        return False
    

    # pop first node from Linked List
    def pop_first(self):
        popped_node = self.head
        self.head = self.head.next
        popped_node.next = None
        return popped_node

    def delete_last_node(self):
        # if the list is empty
        if self.head is None:
            print("Linked list is empty")
            return None
        
        # if only one node
        if self.head.next is None:
            self.head = None
            return
        
        temp = self.head
        while temp.next.next: # traverse to 2nd last node
            temp = temp.next
        temp.next = None # Remove the last node


    # Display items of Linked List
    def display(self):
        """Displays the linked list elements."""
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

    # traversing a singly Linked list 
    def traverse(self):
        current = self.head
        while current is not None:
            print(current.data, end = '  ')
            current = current.next

    
    def remove(self, index):
        current = self.head
        for _ in range(index-1):
            current = current.next
        popped_node = current.next
        




        




        


        

# Example usage        
ll = LinkedList() 
ll.insert_at_beginning(15)
ll.insert_at_beginning(10)
ll.insert_at_beginning(5)
ll.insert_at_end(20)
ll.insert_at_end(30)
ll.insert_at_end(35)
ll.display()
print("before insert")
ll.insert_at_middle(25, 4)
ll.display()

print(".......Traversing Linked list........")
ll.traverse()

print("\n search in Linked List")
result = ll.search(30)
if result:
    print("Item found")
else:
    print("Not found")


print(ll.update_value(3, 100))
ll.display()


# =======================================================

popped_node = ll.pop_first()
print(popped_node)
ll.display()

ll.delete_last_node()
ll.display()

ll.remove()



```
